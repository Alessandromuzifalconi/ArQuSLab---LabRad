```python
"""Driver class for Turbo-V 81-AG gauge controll"""
from __future__ import print_function
import time
import serial
import socket
from serial import Serial
import time
import pyvisa
from pyvisa.constants import StopBits, Parity

class TurboDriver():
    """Driver for Turbo gauge controller"""

    def __init__(self):
        self.ser = serial.Serial(
            port = 'COM5',
            baudrate = 9600,
            parity = serial.PARITY_NONE,
            stopbits = serial.STOPBITS_ONE,
            bytesize = serial.EIGHTBITS,
            timeout = 1, # IMPORTANT, can be lower or higher
            # inter_byte_timeout=0.1 # Alternative
            xonxoff=False
            )

    def read_power(self):
        STX = b'\x02'
        ADDR = b'\x80'
        WIN = b'\x32' + b'\x30' + b'\x32'
        COM = b'\x30'
        ETX = b'\x03'
        CRC = b'\x38' + b'\x33'
        newline = b'\x0A' #\n in hex, D for \r

        command = STX + ADDR + WIN + COM + ETX + CRC + newline
        self.ser.write(command)
        msg = self.ser.readline()
        msg = msg.replace(b"\x80", b"")
        # msg = msg.replace(b"\x0384", b"")
        msg = msg.replace(b"\x02202", b"")
        # msg = msg.replace(b"   \x03E5", b"")
        # msg = msg.decode('utf-8')
        # msg.replace('E', 'e')
        
        msg = msg.decode('utf-8')
        msg = msg[1:7]
        msg = float(msg)
        return msg

    def read_temperature(self):
        STX = b'\x02'
        ADDR = b'\x80'
        WIN = b'\x32' + b'\x30' + b'\x34'
        COM = b'\x30'
        ETX = b'\x03'
        CRC = b'\x38' + b'\x35'
        newline = b'\x0A' #\n in hex, D for \r

        command = STX + ADDR + WIN + COM + ETX + CRC + newline
        self.ser.write(command)
        msg = self.ser.readline()
        msg = msg.replace(b"\x80", b"")
        # msg = msg.replace(b"\x0384", b"")
        msg = msg.replace(b"\x02204", b"")
        msg = msg.replace(b"\x0385", b"")
        # msg = msg.replace(b"0000", b"")
        # msg = msg.decode('utf-8')
        # msg.replace('E', 'e')
        
        msg = msg.decode('utf-8')
        msg = msg[5:7]
        msg = float(msg)
        return msg
        

    def close(self):
        self.ser.close()       

    def read_pressure(self):
        STX = b'\x02'
        ADDR = b'\x80'
        WIN = b'\x32' + b'\x32' + b'\x34'
        COM = b'\x30'
        ETX = b'\x03'
        CRC = b'\x38' + b'\x37'
        newline = b'\x0A' #\n in hex, D for \r

        command = STX + ADDR + WIN + COM + ETX + CRC + newline
        self.ser.write(command)
        msg = self.ser.readline()
        msg = msg.replace(b"\x80", b"")
        msg = msg.replace(b"   \x03E5", b"")
        # msg = msg.decode('utf-8')
        # msg.replace('E', 'e')
        
        msg = msg.decode('utf-8')
        msg = msg[5:12]
        msg = float(msg)
        return msg

    def read_velocity(self):
        STX = b'\x02'
        ADDR = b'\x80'
        WIN = b'\x32' + b'\x32' + b'\x36'
        COM = b'\x30'
        ETX = b'\x03'
        CRC = b'\x38' + b'\x35'
        newline = b'\x0A' #\n in hex, D for \r

        command = STX + ADDR + WIN + COM + ETX + CRC + newline
        self.ser.write(command)
        msg = self.ser.readline()
        msg = msg.replace(b"\x80", b"")
        # msg = msg.replace(b"\x0385", b"")
        # msg = msg.decode('utf-8')
        # msg.replace('E', 'e')
        
        msg = msg.decode('utf-8')
        msg = msg[5:11]
        msg = float(msg)
        msg = msg/60
        return msg

        
    
    
Turbo = TurboDriver()
```
