```python
import numpy as np
n_vector = 3
freq_matrix = np.zeros((n_vector, 3))

for i in range(n_vector):
    values = input(f'Enter three space-separated values for vector {i + 1} (np.arange styled): ')
    freq_matrix[i] = np.array(list(map(float, values.split())))

print(freq_matrix)

vectors_matrix = np.zeros((1, n_vector), dtype=object)
for i in range(n_vector):
    if freq_matrix[i][0] != freq_matrix[i][1]:
        vectors_matrix[0][i] = np.arange(freq_matrix[i][0], freq_matrix[i][1], freq_matrix[i][2])
    elif freq_matrix[i][0] == freq_matrix[i][1]:
        vectors_matrix[0][i] = np.array([freq_matrix[i][0]] * int(freq_matrix[i][2]))
        
vectors_matrix = vectors_matrix.ravel()

joined_string = ''.join(map(str, vectors_matrix))

joined_string = joined_string.replace('[', '').replace(' ', ',').replace(']', ',' ).replace('\n', '')
print(joined_string)

```

```python
inst.write(':list:FREQ ' + joined_string)
```
